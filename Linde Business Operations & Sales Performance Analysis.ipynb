```python
#Total Production Volume 
total_production = df['Production_Volume_Tons'].sum()
print("Total Production Volume: ", total_production, "Tons")
```

```python
#Total Sales Volume
total_sales = df['Sales_Volume_Tons'].sum()
print("Total Sales Volum: ", total_sales, "Tons")
```

```python
#Total Revenue
total_revenue = df['Revenue_INR'].sum()
print("Total Revenue: ₹{:,.0f}".format(total_revenue))
```

```python
#Total Production Cost
total_cost = df['Production_Cost_INR'].sum()
print("Total Production Cost: ₹{:,.0f}".format(total_cost))
```

```python
#Total Profit
total_profit = df['Profit_INR'].sum()
print("Total Profit:  ₹{:,.0f}".format(total_profit))
```

```python
#Average Customer Satisfaction
avg_customer = df['Customer_Satisfaction'].mean()
print("Average Customer Satisfaction: ",round(avg_customer,2))
```

```python
#Average Plant Efficiency
avg_efficiency = df['Plant_Efficiency_Percent'].mean()
print("Avergae Plant Efficiency: ",round(avg_efficiency,2),"%")
```

```python
#Product-wise Revenue
product_revenue = df.groupby('Product')['Revenue_INR'].sum()
print(product_revenue)
```

```python
#Product-wise Profit
product_profit = df.groupby('Product')['Profit_INR'].sum()
print(product_profit)
```

```python
#Plant-wise Revenue
plant_revenue =  df.groupby('Plant')['Revenue_INR'].sum()
print(plant_revenue)
```

```python
#Industry-wise Revenue
industry_revenue = df.groupby('Industry')['Revenue_INR'].sum()
print(industry_revenue)
```

```python
#Region-wise Revenue 
region_revenue = df.groupby('Region')['Revenue_INR'].sum()
print(region_revenue)
```

```python
#Year-wise Revenue
year_revenue = df.groupby('Year')['Revenue_INR'].sum()
print(year_revenue)
```

```python
#Monthly Production 
monthly_production = df.groupby('Month')['Production_Volume_Tons'].sum()
print(monthly_production)
```

```python
correlation = df.corr(numeric_only=True)
print(correlation)
```

```python
#Visualization 1 - Product-wise Revenue 
plt.figure(figsize=(10,6))

product_revenue.plot(kind='bar', color='steelblue')

plt.title("Revenue by Product")

plt.xlabel("Product")

plt.ylabel("Revenue (₹)")

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 2 - Product-wise Profit
plt.figure(figsize=(10,6))

product_profit.plot(kind='bar', color='green')

plt.title("Profit by Product")

plt.xlabel("Product")

plt.ylabel("Profit (₹)")

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 3 - Plant-wise Revenue
plt.figure(figsize=(10,6))

plant_revenue.plot(kind='bar', color='orange')

plt.title("Revenue by Plant")

plt.xlabel("Plant")

plt.ylabel("Revenue")

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 4 - Industry-wise Revenue
plt.figure(figsize=(10,6))

industry_revenue.plot(kind='bar', color='purple')

plt.title("Revenue by Industry")

plt.xlabel("Industry")

plt.ylabel("Revenue")

plt.xticks(rotation=30)

plt.grid(axis='y')

plt.show()
```

```python
#Visualization 5 - Region-wise Revenue
plt.figure(figsize=(8,6))

region_revenue.plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")

plt.title("Revenue Distribution by Region")

plt.show()
```

```python
#Visualization 6 - Monthly Production Trend
plt.figure(figsize=(10,5))

monthly_production.plot(marker='o', linewidth=2)

plt.title("Monthly Production Volume")

plt.xlabel("Month")

plt.ylabel("Production (Tons)")

plt.grid()

plt.show()
```

```python
#Visualization 7 - Revenue vs Production Cost
plt.figure(figsize=(8,5))

plt.scatter(df['Revenue_INR'], df['Production_Cost_INR'])

plt.xlabel("Revenue")

plt.ylabel("Production Cost")

plt.title("Revenue vs Production Cost")

plt.grid()

plt.show()
```

```python
#Visualization 8 - Profit Distribution
plt.figure(figsize=(8,5))

plt.hist(df['Profit_INR'], bins=20)

plt.title("Profit Distribution")

plt.xlabel("Profit")

plt.ylabel("Frequency")

plt.show()
```

```python
#Visualization 9 - Customer Satisfaction
plt.figure(figsize=(8,5))

plt.hist(df['Customer_Satisfaction'], bins=10)

plt.title("Customer Satisfaction")

plt.xlabel("Rating")

plt.ylabel("Frequency")

plt.show()
```

```python
#Visualization 10 - Plant Efficiency
plt.figure(figsize=(8,5))

plt.hist(df['Plant_Efficiency_Percent'], bins=15)

plt.title("Plant Efficiency")

plt.xlabel("Efficiency (%)")

plt.ylabel("Frequency")

plt.show()
```
